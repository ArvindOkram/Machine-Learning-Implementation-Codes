```python
#NAIVE BAYES CLASSIFIER
import pandas as pd

#input data 14x4 table
input_data = {'Outlook':['Rainy','Rainy','Overcast','Sunny','Sunny','Sunny','Overcast',
                         'Rainy','Rainy','Sunny','Rainy','Overcast','Overcast','Sunny'],
              'Humidity':['High','High','High','High','Normal','Normal','Normal',
                          'High','Normal','Normal','Normal','High','Normal','High'],
              'Windy': ['False','True','False','False','False','True','True',
                        'False','False','False','True','True','False','True'],
              'Play Football' : ['No','No','Yes','Yes','Yes','No','Yes',
                                 'No','Yes','Yes','Yes','Yes','Yes','No']}

df = pd.DataFrame(input_data)
print(df)
print("\n")

#Getting each columns individually for calculation of NAIVE BAYES CLASSIFIER
outlook_list = df['Outlook'].tolist()
humidity_list = df['Humidity'].tolist()
windy_list = df['Windy'].tolist()
play_football_list = df['Play Football'].tolist()

class_set = set(play_football_list) 
class_size = len(class_set)

#Counting the number of rows
m = len(play_football_list)
count_yes = count_no = 0
for i in range(m):
    if(play_football_list[i] == 'Yes'):
        count_yes = count_yes+1
    else:
        count_no = count_no+1
        
prob_yes = count_yes/m         #probability of yes
prob_no = count_no/m           #probability of no

"""
                                    TRAINING THE MODEL
"""

#Likelihood of Outlook ie Sunny, Overcast, Rainy
count_outlook_sunny_yes = count_outlook_sunny_no = count_outlook_overcast_yes = 0
count_outlook_overcast_no = count_outlook_rainy_yes = count_outlook_rainy_no = 0

#Loop to count value for probability calculation
for i in range(m):
    if(outlook_list[i] == 'Sunny' and play_football_list[i] == 'Yes'):
        count_outlook_sunny_yes = count_outlook_sunny_yes + 1
    elif(outlook_list[i] == 'Sunny' and play_football_list[i] == 'No'):
        count_outlook_sunny_no = count_outlook_sunny_no + 1
    elif(outlook_list[i] == 'Overcast' and play_football_list[i] == 'Yes'):
        count_outlook_overcast_yes = count_outlook_overcast_yes + 1
    elif(outlook_list[i] == 'Overcast' and play_football_list[i] == 'No'):
        count_outlook_overcast_no = count_outlook_overcast_no + 1
    elif(outlook_list[i] == 'Rainy' and play_football_list[i] == 'Yes'):
        count_outlook_rainy_yes = count_outlook_rainy_yes + 1
    else:
        count_outlook_rainy_no = count_outlook_rainy_no + 1

#Calculation Likelihood of Outlook when Play Football is Yes/No
prob_sunny_yes = count_outlook_sunny_yes+1 / count_yes+class_size
prob_sunny_no = count_outlook_sunny_no+1 / count_no+class_size
prob_overcast_yes = count_outlook_overcast_yes+1 / count_yes+class_size
prob_overcast_no = count_outlook_overcast_no+1 / count_no+class_size
prob_rainy_yes = count_outlook_rainy_yes+1 / count_yes+class_size
prob_rainy_no = count_outlook_rainy_no+1 / count_no+class_size

#Calculating Evidence or Marginal
prob_sunny = prob_sunny_yes + prob_sunny_no
prob_overcast = prob_overcast_yes + prob_overcast_no
prob_rainy =  prob_rainy_yes + prob_rainy_no

#Calculation Likelihood of Yes and No for all possible outlooks
likelihood_of_yes_given_sunny =  prob_sunny_yes * prob_yes / prob_sunny
likelihood_of_no_given_sunny =  prob_sunny_no * prob_no / prob_sunny
likelihood_of_yes_given_overcast = prob_overcast_yes * prob_yes / prob_overcast
likelihood_of_no_given_overcast = prob_overcast_no * prob_no / prob_overcast
likelihood_of_yes_given_rainy = prob_rainy_yes * prob_yes / prob_rainy
likelihood_of_no_given_rainy = prob_rainy_no * prob_no / prob_rainy

#Storing result for Likelihood of Yes for all possible of outlooks
dic_outlook_yes = {'Sunny' : likelihood_of_yes_given_sunny,
                   'Overcast' : likelihood_of_yes_given_overcast,
                   'Rainy' : likelihood_of_yes_given_rainy}

#Storing result for Likelihood of No for all possible of outlooks
dic_outlook_no = {'Sunny' : likelihood_of_no_given_sunny,
                  'Overcast' : likelihood_of_no_given_overcast,
                  'Rainy' : likelihood_of_no_given_rainy}
#____________________________________________________________________________________________________
#Likelihood of Humidity ie High, Normal
count_humidity_high_yes = count_humidity_high_no = count_humidity_normal_yes = count_humidity_normal_no = 0

#Loop to count value for probability calculation
for i in range(m):
    if (humidity_list[i] =='High' and play_football_list[i] == 'Yes'):
        count_humidity_high_yes = count_humidity_high_yes + 1
    elif (humidity_list[i] =='High' and play_football_list[i] == 'No'):
        count_humidity_high_no = count_humidity_high_no + 1
    elif (humidity_list[i] =='High' and play_football_list[i] == 'Yes'):
        count_humidity_normal_yes = count_humidity_normal_yes + 1
    else:
        count_humidity_normal_no = count_humidity_normal_no + 1 

#Calculation Likelihood of Humidity when Play Football is Yes/No
prob_high_yes = count_humidity_high_yes+1 / count_yes+class_size
prob_high_no = count_humidity_high_no+1 / count_no+class_size
prob_normal_yes = count_humidity_normal_yes+1 / count_yes+class_size
prob_normal_no = count_humidity_normal_no+1 / count_no+class_size

#Calculating Evidence or Marginal
prob_high = prob_high_yes + prob_high_no
prob_normal = prob_normal_yes + prob_normal_no

#Calculation Likelihood of Yes and No for all possible of humidities
likelihood_of_yes_given_high =  prob_high_yes * prob_yes / prob_high
likelihood_of_no_given_high =  prob_high_no * prob_no / prob_high
likelihood_of_yes_given_normal = prob_normal_yes * prob_yes / prob_normal
likelihood_of_no_given_normal = prob_normal_no * prob_no / prob_normal

#Storing result for Likelihood of Yes for all possible of humidities
dic_humidity_yes = {'High' : likelihood_of_yes_given_high,
                    'Normal' : likelihood_of_yes_given_normal}

#Storing result for Likelihood of No for all possible of humidities
dic_humidity_no = {'High' : likelihood_of_no_given_high,
                    'Normal' : likelihood_of_no_given_normal}
#________________________________________________________________________________________________________
#Likelihood of Windy ie True, False
count_windy_true_yes = count_windy_true_no = count_windy_false_yes = count_windy_false_no = 0

#Loop to count value for probability calculation
for i in range(m):
    if (windy_list[i] == 'True' and play_football_list[i] == 'Yes'):
        count_windy_true_yes = count_windy_true_yes + 1
    elif (windy_list[i] == 'True' and play_football_list[i] == 'No'):
        count_windy_true_no = count_windy_false_no + 1
    elif (windy_list[i] == 'False' and play_football_list[i] == 'Yes'):
        count_windy_false_yes = count_windy_false_yes + 1
    else:
        count_windy_false_no = count_windy_false_no + 1

#Calculation Likelihood of Windy when Play Football is Yes/No
prob_true_yes = count_windy_true_yes+1/count_yes+class_size
prob_true_no = count_windy_true_no+1/count_no+class_size
prob_false_yes = count_windy_false_yes+1/count_yes+class_size
prob_false_no = count_windy_false_no+1/count_no+class_size

#Calculating Evidence or Marginal
prob_true = prob_true_yes + prob_true_no
prob_false = prob_false_yes + prob_false_no

#Calculation Likelihood of Yes and No for all possible of windy
likelihood_of_yes_given_true =  prob_true_yes * prob_yes / prob_true
likelihood_of_no_given_true =  prob_true_no * prob_no / prob_false
likelihood_of_yes_given_false =  prob_false_yes * prob_yes / prob_false
likelihood_of_no_given_false =  prob_false_no * prob_no / prob_false

#Storing result for Likelihood of Yes for all possible of windy
dic_windy_yes = {'True' : likelihood_of_yes_given_true,
                 'False' : likelihood_of_yes_given_false}

#Storing result for Likelihood of No for all possible of windy
dic_windy_no = {'True' : likelihood_of_no_given_true,
                'False' : likelihood_of_no_given_false}

"""
                                    TESTING THE MODEL
"""


#test data
test_outlook = input('Enter data for Outlook ie Sunny/Overcast/Rainy : ')
test_humidity = input('Enter data for Humidity ie High/Normal: ')
test_windy = input('Enter data for Windy ie True/False: ')

result_outlook_yes = dic_outlook_yes[test_outlook]
result_humidity_yes = dic_humidity_yes[test_humidity]
result_windy_yes = dic_windy_yes[test_windy]
result_outlook_no = dic_outlook_no[test_outlook]
result_humidity_no = dic_humidity_no[test_humidity]
result_windy_no = dic_windy_no[test_windy]

#Calculating likelihood of yes that day
naive_bayes_result_yes = result_outlook_yes * result_humidity_yes * result_windy_yes

#Calculating likelihood of no that day
naive_bayes_result_no = result_outlook_no * result_humidity_no * result_windy_no

if(naive_bayes_result_yes + naive_bayes_result_yes == 0):
    #Bottleneck case of Naive Bayes Classifier
    print("'ZERO FREQUENCY PROBLEM' occured, a drawback of Naive Bayes Classifier")
else:
    prob_yes_that_day = naive_bayes_result_yes / (naive_bayes_result_yes + naive_bayes_result_no)
    prob_no_that_day = naive_bayes_result_no / (naive_bayes_result_yes + naive_bayes_result_no)
    print("Probability of playing football that day : {}".format(prob_yes_that_day))
    print("Probability of not playing football that day : {}".format(prob_no_that_day))        

    
```
